```python
import seaborn as sns
#Pandas library is going to be used for data manipulation 
import pandas as pd 
#Matplotlib library is going to be used for data visualization
from matplotlib import pyplot as plt
#df(Dtataframe) is going to hold the csv file(Final.csv)
df=pd.read_csv('Final.csv')
#Head is going to display the first 5 rows of the csv file
df.head(5)
```

```python
#Presenting and indentifying the data type of the dataframe
df.dtypes
```

```python
#Droping the un-wanted coulmns
df=df.drop(['Unnamed: 0','Incedents'],axis=1)
#Displaying the first 3 rows of the csv file 
df.head(3)
```

```python
#Changing the Death column fromat to only dispaly the first number not the (passengers-crew )
df['Death'] = df['Death'].str.replace(r" \(.*\)","")
#Displaying the first 2 rows of the csv file 
df.head(2)
```

```python
#Changing the Aboard column fromat to only dispaly the first number not the (passengers-crew )
df['Aboard'] = df['Aboard'].str.replace(r" \(.*\)","")
#Displaying the first 2 rows of the csv file 
df.head(2)
```

```python
#Saving the changes made on the Final.csv file and saving it in a new file called Private.csv
df.to_csv('Private.csv')
```

```python
#df1(Dateframe1) is going to hold the Private.csv file
df1=pd.read_csv('Private.csv')
#Dispalyiing the Private file
df1
```

```python
#To check the new data types in the updated csv file (Private.csv)
df1.dtypes
```

```python
#Converting the Dtae from an object to python date
df1['Date']=pd.to_datetime(df1['Date'])
#Displaying the Date column
df1['Date']
```

```python
#Showing the year only from the date
Year=pd.DatetimeIndex(df1['Date']).year
#Displaying the year 
Year
```

```python
#Grouping the year and the date count 
Group=df1.groupby([Year])[['Date']].count()
#Selecting the graph size
plt.figure(figsize=(10,7))
#Ploting the graph according to the Group fuction 
plt.plot(Group.index,'Date',data=Group,color="Black",linewidth=1.5)
#Labeling the x-axsis
plt.xlabel("Year",fontsize=13)
#Labeling the y-axsis
plt.ylabel("Accidents",fontsize=13)
#The title of the graph
plt.title("Total Accidents",fontsize=25)
#Showing the graph
plt.show()
```

```python
#Groupby the year with the sum of the Aboard
Group2=df1.groupby([Year])[["Aboard"]].sum()
Group3=df1.groupby([Year])[["Death"]].sum()
#Selecting the graph size
plt.figure(figsize=(10,7))
#Ploting the graph according to the Group2(Aboard) and Group3(Death) function 
plt.plot(Group2.index,'Aboard',data=Group2,color="Blue",linewidth=1)
plt.plot(Group3.index,'Death',data=Group3,color="Red",linewidth=1)
#Showing which line is (Aboard) and which is (Death)
plt.legend()
#Labeling the x-axsis
plt.xlabel("Year",fontsize=13)
#Labeling the y-axsis
plt.ylabel("Aboard",fontsize=13)
#The title of the graph
plt.title("Comparison of Death and Aboard",fontsize=25)
#Showing the graph
plt.show()
```

```python
#Groupby the Ac Type and counting the Ac Type
go=df1.groupby(["AC Type"])[["AC Type"]].count()
#Renaming the Ac Type to Type 
Graph = go.rename(columns={"AC Type": "Type"})
#Sorting The Type into ascending order to display the  top 10
Graph1 = Graph.sort_values(by='Type', ascending=False).head(10)
#The figure size
plt.figure(figsize=(13,7))
# Designing the graph
sns.barplot(y=Graph1.index, x="Type",data=Graph1, color='black')
#Labeling the x-axsis
plt.xlabel('Total Accidents', fontsize=15)
#Labeling the y-axsis
plt.ylabel('AC Type', fontsize=15)
#Labeling the title
plt.title('Accidents caused by most common AC types', loc='center', fontsize=18)
#Showing the graph
plt.show()
```

```python
#Importing pandas library
import pandas as pd 
#Matplotlib library is used to plot graphs 
from matplotlib import pyplot as plt
#Dataframe that will hold on the csv file 
df=pd.read_csv('Final.csv') 
#Seaborn library is used for high-level graphs
import seaborn as sns
#Displaying the first 3 rows of the csv file 
df.head(3)
```

```python
#Coverting the data from object to python date to be used latter 
df['Date']=pd.to_datetime(df['Date']) 
#Showing the date after changing the data type
df['Date'] 
```

```python
#Extracting the year only from the date 
df['Date'].dt.year 
```

```python
#Adding the extracted year into the dataframe (Final) as a new column 
df['Year']=df['Date'].dt.year
```

```python
#Displaying the first 5 rows
df.head(5) 
```

```python
#Changing the date format from('%Y-%m-%d') to ('%d/%b/%Y')
df['Date']= df['Date'].dt.strftime('%d/%b/%Y') 
#Displaying the first 5 rows
df.head(5)
```

```python
#Changing the death column to only dissplay the total number of death and adding it to the dataframe(Final)as a new column
df['Number_deaths']=df['Death'].str.replace(r"\(.*\)","")
#Displaying the first 5 rows
df.head(5) 
```

```python
#Saving every step done above into a new csv file to be used for visualiztion 
df.to_csv('Graphs.csv') 
```

```python
#Reading the saved dataframe (Graphs)
df2=pd.read_csv('Graphs.csv') 
#Displaying the dataframe(Graphs)
df2 
```

```python
#Removing unnecessary columns 
df2.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis = 1) 
```

```python
# Grouping the Year column with its count to be used for a garph and naming it Figure1
Figure1=df2.groupby(['Year'])[['Year']].count() 
#Displaying the groupby 
Figure1 
```

```python
#Grouping the Year with the Number_deaths sum to be used for a garph and naming it Figure2
Figure2=df2.groupby(['Year'])[['Number_deaths']].sum()
#Displaying the groupby 
Figure2
```

```python
#Grouping the location with the number of deaths sum to be used for a garph and naming it Figure3
Figure3=df2.groupby('Location')[['Number_deaths']].sum()
#Displaying the groupby 
Figure3
```

```python
#Grouping the location with its count to be displayed in a graph and naming it Figure4
Figure4=df2.groupby(['Location'])[['Location']].count()
#Displaying the groupby
Figure4
```

```python
# Displaying the groupby and naming it Graph1
Graph1 = Figure1 
# Creating a font_style to be used for the graph 1
font1 = {'family':'Arial','color':'Black','size':15} 
#Adjusting the graphs size
plt.figure(figsize=(12,5)) 
# Choosing the style of the graph 
plt.style.use('default') 
# Ploting the operation of the graph 
plt.plot(Graph1.index,'Year', data=Graph1, color='blue',marker='o',linewidth=2)
#Deciding to add a grid for my graph
plt.grid()  
#Naming the x-axsis
plt.xlabel('Year',loc='center', fontdict = font1) 
#Naming the y-axis
plt.ylabel('Number of incedents',loc='center', fontdict = font1) 
 #Naming the graph title to give an idea
plt.title('Incedents per year', loc='Center',fontname='Comic Sans MS', fontsize=20)
#Showing the graph with everything addjusted in the graph
plt.show() 
# Displaying the groupby and naming it Graph2
Graph2=Figure2 
# The created font for graph2
font2 = {'family':'Arial','color':'Black','size':15} 
#Adjusting the graphs size
plt.figure(figsize=(12,5))  
# Choosing the style of the graph 
plt.style.use('default')
# Ploting the operation of the graph 
plt.plot(Graph2.index, 'Number_deaths', data=Graph2, color="black",marker='o', linewidth=2) 
#Deciding to add a grid for my graph 
plt.grid() 
#Naming the x-axsis
plt.xlabel('Year', loc='center',fontdict = font2)
#Naming the y-axis
plt.ylabel('Number of deaths',loc='center', fontdict = font2) 
#Naming the graph title to give an idea
plt.title('Deaths per year', loc='center',fontname='Comic Sans MS', fontsize=20) 
#Showing the graph with everything addjusted in the graph
plt.show()
#Displaying the groupby and naming it Graph3
Graph3 = Figure3
# The created font for graph three 
font3 = {'family':'Arial','color':'Black','size':15} 
#Position will hold the fuction that will sort the Location according to the Number_deaths
Position=Graph3.sort_values(by='Number_deaths', ascending=False)
#Amount will display the twenty of the sorted data(Position)
Amount = Position.head(20)
#Adjusting the graphs size
plt.figure(figsize=(12,5))
# Ploting the operation of the graph 
sns.barplot(y=Amount.index, x="Number_deaths", data=Amount, palette="viridis")
#Naming the x-axsis
plt.xlabel('Number of deaths ',fontdict = font3)
#Naming the y-axsis
plt.ylabel('Location',fontdict = font3)
#Naming the graph title to give an idea
plt.title('Location per Deaths', loc='Center',fontname='Comic Sans MS',fontsize=20)
#Showing the graph with everything addjusted in the graph
plt.show()
#Displaying the groupby and naming it Graph4
Graph4=Figure4
# The created font for graph four
font4 = {'family':'Arial','color':'Black','size':15} 
#Renaming the Location into Locations so it could be used
Graph = Graph4.rename(columns={"Location": "Locations"})
#Fig will hold the fuction that will sort the Locations according to the Number of Incedents
Fig = Graph.sort_values(by='Locations', ascending=False).head(20)
#Adjusting the graphs size
plt.figure(figsize=(12,5))
# Ploting the operation of the graph 
sns.barplot(y=Fig.index, x="Locations", data=Fig,palette="icefire" )
#Naming the x-axsis
plt.xlabel('Total Incedents', fontdict = font4)
#Naming the y-axsis
plt.ylabel('Locations', fontdict = font4)
#Naming the graph title to give an idea
plt.title('Locations per Incedents', loc='Center',fontname='Comic Sans MS',fontsize=20)
#Showing the graph with everything addjusted in the graph
plt.show()
```

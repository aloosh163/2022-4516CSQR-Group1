```python
import pandas as pd
from matplotlib import pyplot as plt 
import seaborn as sns
df=pd.read_csv('Final.csv')
df
```

```python
df['Death']=df['Death'].str.replace(r"\(.*\)","")
```

```python
df['Aboard']=df['Aboard'].str.replace(r"\(.*\)","")
```

```python
df.head(4)
```

```python
df['Death']=df['Death'].astype(float)
df['Aboard']=df['Aboard'].astype(float)
df.head(5)
```

```python
df.dtypes
```

```python
df['Survive']=df['Aboard']-df['Death']
df.head(3)
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['Year']=df['Date'].dt.year
```

```python
df.head(5)
```

```python
todo=df.groupby(['Year'])[['Death']].sum()
todo
```

```python
got=df.loc[:,['Year','Death','Survive']].groupby(['Year']).sum()
plot=got.plot.bar(stacked=True,figsize=(20,12))
plt.style.use('seaborn-white')
plt.title('Yearly Survived and Death',fontsize=20)
plt.xlabel('Year')
plt.ylabel("Number of people Aboard")
plt.show()
```

```python
routes=df['Route']
routes=routes.dropna().str.lower().str.replace(" ", "").str.split("-")
routes=routes[routes.str.len()==2]
dept2Dest=pd.DataFrame({'Departure': routes.str.get(0), 'Destination': routes.str.get(1)})
```

```python
departure = dept2Dest.groupby(['Destination']).count().sort_values(by="Departure", ascending=False).head(10)
Dep = departure.plot.bar(figsize=(10,5), title="Top Ten Most Dangerous City for Departure", fontsize=13)
plt.style.use('ggplot')
Dep.set_xlabel("City", fontsize=15)
Dep.set_ylabel("Number of Crash", fontsize=10)
```

```python
destination = dept2Dest.groupby(['Departure']).count().sort_values(by="Destination", ascending=False).head(10)
plot8 = destination.plot.bar(figsize=(10,5), title="Top Ten Most Dangerous City for Destination", fontsize=13,color='Blue')
plt.style.use('ggplot')
plot8.set_xlabel("City", fontsize=15)
plot8.set_ylabel("Number of Crash", fontsize=10)
```

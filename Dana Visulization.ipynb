```python
import pandas as pd
Start=pd.read_csv('Final.csv')
Start
```

```python
Start['Toatal death']=Start['Death'].str.replace(r"\(.*\)","")
```

```python
Start['Toatal death']=Start['Toatal death'].astype(float)
Start.head(5)
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
number1=Start.groupby('Operator')[['Operator']].count()
plt.figure(figsize=(13,7))
Graph = number1.rename(columns={"Operator": "Total_Operator"})
Graph1 = Graph.sort_values(by='Total_Operator', ascending=False).head(15)
sns.barplot(y=Graph1.index, x="Total_Operator", data=Graph1, color='pink')
plt.xlabel('Total', fontsize=15)
plt.ylabel('Operator_type', fontsize=15)
plt.title('Total of operators', loc='right', fontsize=15)
plt.show()
```

```python
number2=Start.groupby('Operator')[['Toatal death']].sum()
Graph1 = number2.sort_values(by='Toatal death', ascending=False).head(15)
plt.figure(figsize=(13,7))
sns.barplot(y=Graph1.index, x="Toatal death", data=Graph1, color='pink')
plt.xlabel('Total_Deaths', fontsize=15)
plt.ylabel('Operator_Type', fontsize=15)
plt.title('Total_Deaths by Operator ', loc='right', fontsize=15)
plt.show()
```
